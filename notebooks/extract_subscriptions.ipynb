```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../presencas/inscricoes.csv")
df.head()
```

```python
df.columns
```

```python
idx_pt = df["Idioma/Language"] == "Português"
idx_es = df["Idioma/Language"] == "Espanhol"
idx_en = df["Idioma/Language"] == "Inglês"
```

```python
columns = [
    'Timestamp', 
    'Idioma/Language', 
    'E-mail:', 
    'Nome completo:',
    'Nível de Escolaridade:', 
    'Localidade atual:',
    'Aceitar termo de consentimento '
]

ncolumns = [
    "datetime",
    "lang",
    "email",
    "name",
    "institution",
    "level",
    "local",
    "about",
    "minicourse",
    "hackathon",
    
]
```

```python
from calendar import c


columns_en = [
    'Timestamp', 
    'Idioma/Language', 
    'E-mail', 
    'Full name',
    'Institution/Company', 'Education Level',
    'Current Location\nCity and country where you currently reside. \nFor example: São Paulo - Brasil.',
    'How did you hear about WorCAP 2022?',
    'Desejo me inscrever no(s) MINICURSO(S):',
    'Unnamed: 21'
]

columns_es = [
    'Timestamp', 
    'Idioma/Language', 
    'E-mail.1', 
    'Nombre Completo',
    'Institución/Empresa', 
    'Nivel de Formación',
    'Localidad actual\nCiudad y País donde reside actualmente. \nEjemplo: São Paulo - Brasil.',
    '¿Como se entero del WorCAP 2022?',
    'Desejo me inscrever no(s) MINICURSO(S):',
    'Unnamed: 21'
]

columns_pt = [
    'Timestamp', 
    'Idioma/Language', 
    'E-mail:', 
    'Nome completo:',
    'Instituição/Empresa:', 
    'Nível de Escolaridade:', 
    'Localidade atual:',
    'Como ficou sabendo do WorCAP 2022?',
    'Desejo me inscrever no(s) MINICURSO(S):',
    'Unnamed: 21'
]
```

```python
df_pt = df[idx_pt][columns_pt]
df_pt.columns = ncolumns
df_es = df[idx_es][columns_es]
df_es.columns = ncolumns
df_en = df[idx_en][columns_en]
df_en.columns = ncolumns
new_df = pd.concat([df_pt, df_es, df_en], axis=0)
new_df.head()
```

```python
# alter to 0 hackathon column if NaN else 1
new_df["hackathon"] = new_df["hackathon"].apply(lambda x: 0 if pd.isna(x) else 1)
new_df['sub_mc1'] = 0 
new_df['sub_mc2'] = 0
new_df['sub_mc3'] = 0
new_df['sub_mc4'] = 0
for i, row in new_df.iterrows():
    if not pd.isna(row["minicourse"]):
        mcs = []
        
        for mc in row["minicourse"].split(","):
            acron = mc.split()[0].strip()
            if acron == "MC1":
                new_df.loc[i, "sub_mc1"] = 1
            elif acron == "MC2":
                new_df.loc[i, "sub_mc2"] = 1
            elif acron == "MC3":
                new_df.loc[i, "sub_mc3"] = 1
            elif acron == "MC4":
                new_df.loc[i, "sub_mc4"] = 1
new_df
```

```python
# Check participants of pl01

temp = pd.read_csv("../presencas/pl01.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl01'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl02
temp = pd.read_csv("../presencas/pl02.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl02'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl03
temp = pd.read_csv("../presencas/pl03.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl03'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl04
temp = pd.read_csv("../presencas/pl04.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl04'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl05
temp = pd.read_csv("../presencas/pl05.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl05'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl06
temp = pd.read_csv("../presencas/pl06.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl06'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl07
temp = pd.read_csv("../presencas/pl07.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl07'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl08
temp = pd.read_csv("../presencas/pl08.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl08'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of pl09
temp = pd.read_csv("../presencas/pl09.csv", names=['datetime', 'email'], skiprows=1)
new_df['pl09'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of st01
temp = pd.read_csv("../presencas/st01.csv", names=['datetime', 'email'], skiprows=1)
new_df['st01'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of st02
temp = pd.read_csv("../presencas/st02.csv", names=['datetime', 'email'], skiprows=1)
new_df['st02'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of mc01
temp = pd.read_csv("../presencas/mc01.csv", names=['datetime', 'email'], skiprows=1)
new_df['mc01'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of mc02
temp = pd.read_csv("../presencas/mc02.csv", names=['datetime', 'email'], skiprows=1)
new_df['mc02'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of mc03
temp = pd.read_csv("../presencas/mc03.csv", names=['datetime', 'email'], skiprows=1)
new_df['mc03'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

# Check participants of mc04
temp = pd.read_csv("../presencas/mc04.csv", names=['datetime', 'email'], skiprows=1)
new_df['mc04'] = new_df['email'].apply(lambda x: 1 if x in temp['email'].values else 0)

new_df
```

```python
columns = [
    'name', 'email', 'hackathon', 'sub_mc1', 'sub_mc2', 'sub_mc3', 'sub_mc4', 
    'pl01', 'pl02', 'pl03', 'pl04', 'pl05', 'pl06', 'pl07', 'pl08', 'pl09', 
    'st01', 'st02', 'mc01', 'mc02', 'mc03', 'mc04'
]
fdf = new_df[columns]
fdf.head()
```

```python
# compute percentage of participation
fdf['part_evento'] = fdf.iloc[:, 7:-7].sum(axis=1) / fdf.iloc[:, 7:-7].shape[1]
fdf['cert_evento'] = fdf['part_evento'].apply(lambda x: 1 if x > 0 else 0)
fdf['cert_mc01'] = fdf['mc01'].apply(lambda x: 1 if x > 0 else 0)
fdf['cert_mc02'] = fdf['mc02'].apply(lambda x: 1 if x > 0 else 0)
fdf['cert_mc03'] = fdf['mc03'].apply(lambda x: 1 if x > 0 else 0)
fdf['cert_mc04'] = fdf['mc04'].apply(lambda x: 1 if x > 0 else 0)
fdf
```
